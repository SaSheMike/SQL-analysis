```python
def longest(strings: list):
  longest = ""
  for string in strings:
    if len(string) > len(longest):
      longest = string
    else:
      continue
  return longest

strings = ["hi", "hiya", "hello", "howdydoody", "hi there"]
print(longest(strings))
```
